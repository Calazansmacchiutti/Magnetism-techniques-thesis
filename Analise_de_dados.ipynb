```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Supondo que os DataFrames já foram carregados anteriormente

# Lista de DataFrames
dfs = [Or1, Or2, Or3, NdBa001, NdBappd]

# Cores para cada DataFrame
cores = ['red', 'blue', 'green', 'purple', 'orange']

# Títulos para cada plot
titulos = ['Warwick Orient1 - 1kOe', 'Warwick Orient2 - 1kOe', 'Warwick Orient3 - 1kOe', 'CBPF Orient_001 - 100Oe', 'CBPF Orient_ppd001 - 100Oe']

# Janela para a média móvel na derivada
window_size = 100

# Criar o plot
plt.figure(figsize=(10, 6))  # Ajustar o tamanho da figura

for df, cor, titulo in zip(dfs, cores, titulos):
    # Filtrando os dados conforme os limites de 'Temperature (K)'
    df_filtered = df[(df['Temperature (K)'] > 2) & (df['Temperature (K)'] < 250)]
    if 'Temperature (K)' in df_filtered.columns:
        y_col = 'chi' if 'chi' in df_filtered.columns else 'emu/g' if 'emu/g' in df_filtered.columns else None
        if y_col and not df_filtered.empty:
            # Calculando a derivada no DataFrame filtrado
            derivative = np.gradient(df_filtered[y_col], df_filtered['Temperature (K)'])

            # Aplicando a média móvel na derivada
            smoothed_derivative = np.convolve(derivative, np.ones(window_size)/window_size, mode='same')

            # Evitar divisão por zero na normalização
            max_derivative = np.max(np.abs(smoothed_derivative))
            if max_derivative == 0:
                max_derivative = 1  # Para evitar divisão por zero

            # Normalizando a derivada suavizada
            derivative_normalized = smoothed_derivative / max_derivative

            plt.plot(df_filtered['Temperature (K)'], derivative_normalized, label=f"{titulo}", color=cor)
        else:
            print(f"A coluna {y_col} não foi encontrada em {titulo} ou DataFrame está vazio.")
    else:
        print(f"A coluna 'Temperature (K)' não foi encontrada em {titulo}")

# Adicionar legendas e títulos
plt.xlabel('Temperature (K)')
plt.ylabel('Normalized Smoothed Derivative')
plt.title('Normalized Smoothed Derivative of Data Across Temperature')
plt.legend()

plt.xlim(4, 180)

# Adicionar linha de grade
plt.grid(True)

# Adicionar linhas nos eixos X=0 e Y=0
plt.axhline(0, color='black', linewidth=0.8)  # Linha horizontal em Y=0
plt.axvline(0, color='black', linewidth=0.8)  # Linha vertical em X=0

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.signal import argrelextrema

# Supondo que os DataFrames já foram carregados e limpos

# Lista de DataFrames
dfs = [Or1, Or2, Or3, NdBa001, NdBappd]

# Cores para cada DataFrame
cores = ['red', 'blue', 'green', 'purple', 'orange']

# Títulos para cada plot
titulos = ['Warwick Orient1 - 1 kOe', 'Warwick Orient2 - 1 kOe', 'Warwick Orient3 - 1 kOe', 'CBPF Orient_001 - 100 Oe', 'CBPF Orient_ppd001 - 100 Oe']

# Janela para a média móvel na derivada
window_size = 50

# Ordem para detecção de máximos e mínimos locais
extrema_order = 300

# Criar o plot
plt.figure(figsize=(10, 6))  # Ajustar o tamanho da figura

for df, cor, titulo in zip(dfs, cores, titulos):
    if 'Temperature (K)' in df.columns:
        y_col = 'chi' if 'chi' in df.columns else 'emu/g' if 'emu/g' in df.columns else None
        if y_col:
            # Filtrando os dados conforme os limites de 'Temperature (K)'
            df_filtered = df[(df['Temperature (K)'] > 5) & (df['Temperature (K)'] < 250)].dropna()

            if not df_filtered.empty:
                # Calculando a derivada no DataFrame filtrado
                derivative = np.gradient(df_filtered[y_col], df_filtered['Temperature (K)'])

                # Aplicando a média móvel na derivada
                smoothed_derivative = np.convolve(derivative, np.ones(window_size)/window_size, mode='same')

                # Encontrar máximos e mínimos locais na derivada suavizada
                max_indices = argrelextrema(smoothed_derivative, np.greater, order=extrema_order)[0]
                min_indices = argrelextrema(smoothed_derivative, np.less, order=extrema_order)[0]

                # Evitar divisão por zero na normalização
                max_derivative = np.max(np.abs(smoothed_derivative))
                if max_derivative == 0:
                    max_derivative = 1  # Para evitar divisão por zero

                # Normalizando a derivada suavizada
                derivative_normalized = smoothed_derivative / max_derivative

                plt.plot(df_filtered['Temperature (K)'], derivative_normalized, label=f"{titulo}", color=cor)
                plt.scatter(df_filtered['Temperature (K)'].iloc[max_indices], derivative_normalized[max_indices], color='black', marker='x', label='Max Locais' if titulo == 'Warwick Orient1 - 1kOe' else "")
                plt.scatter(df_filtered['Temperature (K)'].iloc[min_indices], derivative_normalized[min_indices], color='black', marker='o', label='Min Locais' if titulo == 'Warwick Orient1 - 1kOe' else "")

                # Imprimir os pontos de máximo e mínimo locais
                print(f"Máximos Locais para {titulo}:")
                for i in max_indices:
                    print(f"Temperature (K): {df_filtered['Temperature (K)'].iloc[i]}, Derivada Normalizada: {derivative_normalized[i]}")

                print(f"Mínimos Locais para {titulo}:")
                for i in min_indices:
                    print(f"Temperature (K): {df_filtered['Temperature (K)'].iloc[i]}, Derivada Normalizada: {derivative_normalized[i]}")
            else:
                print(f"O DataFrame {titulo} está vazio após remoção de NaNs.")
        else:
            print(f"A coluna {y_col} não foi encontrada em {titulo}")
    else:
        print(f"A coluna 'Temperature (K)' não foi encontrada em {titulo}")

# Adicionar legendas e títulos
plt.xlabel('Temperature (K)')
plt.ylabel('Normalized Smoothed Derivative')
plt.title('ZFC - NdBaMnO Warwick vs CBPF - 1st derivative smoothed')
plt.legend()

# Adicionar linha de grade
plt.grid(True)

# Adicionar linhas nos eixos X=0 e Y=0
plt.axhline(0, color='black', linewidth=0.8)  # Linha horizontal em Y=0
plt.axvline(0, color='black', linewidth=0.8)  # Linha vertical em X=0

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Supondo que os DataFrames já foram carregados e limpos

# Lista de DataFrames
dfs = [Or1, Or2, Or3]

# Limpeza de todos os DataFrames para remover linhas com qualquer NaN
dfs = [df.dropna() for df in dfs]  # Aplica dropna a cada DataFrame na lista

# Cores para cada DataFrame
cores = ['red', 'blue', 'green', 'purple', 'orange']

# Títulos para cada plot
titulos = ['Warwick Orient1 - 1kOe', 'Warwick Orient2 - 1kOe', 'Warwick Orient3 - 1kOe', 'CBPF Orient_001 - 100Oe', 'CBPF Orient_ppd001 - 100Oe']

# Criar o plot
plt.figure(figsize=(10, 6))  # Ajustar o tamanho da figura

for df, cor, titulo in zip(dfs, cores, titulos):
    if 'Temperature (K)' in df.columns:
        y_col = 'chi' if 'chi' in df.columns else 'emu/g' if 'emu/g' in df.columns else None
        if y_col:
            # Filtrando a região de 300 a 400 do eixo X
            df_filtered = df[(df['Temperature (K)'] > 350) & (df['Temperature (K)'] < 400)].dropna()

            if not df_filtered.empty:
                # Inverter os dados de Temperature (K) e y_col
                inverted_temp = 1 / df_filtered['Temperature (K)']
                y_data = df_filtered[y_col]

                # Realizar o ajuste linear
                slope, intercept, r_value, p_value, std_err = linregress(inverted_temp, y_data)

                # Plotar os dados invertidos
                plt.scatter(inverted_temp, y_data, label=f"{titulo} ({y_col})", color=cor)

                # Plotar a linha de ajuste linear
                plt.plot(inverted_temp, slope * inverted_temp + intercept, color=cor, linestyle='--', label=f"Fit {titulo} (slope: {slope:.2f})")

                # Imprimir os parâmetros do ajuste linear
                print(f"Ajuste Linear para {titulo}: slope = {slope}, intercept = {intercept}")
            else:
                print(f"O DataFrame {titulo} está vazio após remoção de NaNs.")
        else:
            print(f"A coluna {y_col} não foi encontrada em {titulo}")
    else:
        print(f"A coluna 'Temperature (K)' não foi encontrada em {titulo}")

# Adicionar legendas e títulos
plt.xlabel('1 / Temperature (K)')
plt.ylabel('Value')
plt.title('Inverted Data with Linear Fit (Region 300-400K)')
plt.legend()

# Adicionar linha de grade
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Supondo que os DataFrames já foram carregados e limpos

# Lista de DataFrames
dfs = [Or1, Or2, Or3]

# Limpeza de todos os DataFrames para remover linhas com qualquer NaN
dfs = [df.dropna(subset=['Moment (emu)', 'Magnetic Field (Oe)', 'Temperature (K)']) for df in dfs]  # Aplica dropna a cada DataFrame na lista

# Cores para cada DataFrame
cores = ['red', 'blue', 'green']

# Títulos para cada plot
titulos = ['Warwick Orient1 - 1 kOe', 'Warwick Orient2 - 1 kOe', 'Warwick Orient3 - 1 kOe']

# Criar o plot
plt.figure(figsize=(10, 6))  # Ajustar o tamanho da figura

for df, cor, titulo in zip(dfs, cores, titulos):
    if 'Temperature (K)' in df.columns and 'Moment (emu)' in df.columns and 'Magnetic Field (Oe)' in df.columns:
        # Criar a nova coluna conforme a fórmula fornecida
        df['new_col'] = df['Moment (emu)'] * 330.6576 / (0.0072 * df['Magnetic Field (Oe)'])

        # Verificar se a nova coluna foi criada corretamente
        if df['new_col'].isna().all():
            print(f"A nova coluna 'new_col' em {titulo} está vazia após a criação.")
            continue

        # Filtrar a região de 300 a 400 do eixo X (Temperatura)
        df_filtered = df[(df['Temperature (K)'] > 300) & (df['Temperature (K)'] < 400)].dropna(subset=['new_col'])

        if df_filtered.empty:
            print(f"O DataFrame {titulo} está vazio após remoção de NaNs.")
            continue

        # Inverter os dados da nova coluna
        y_data = 1 / df_filtered['new_col']
        x_data = df_filtered['Temperature (K)']

        # Verificar se os dados invertidos são válidos
        if y_data.isna().all() or x_data.isna().all():
            print(f"Os dados invertidos em {titulo} contêm valores NaN.")
            continue

        # Realizar o ajuste linear
        slope, intercept, r_value, p_value, std_err = linregress(x_data, y_data)

        # Plotar os dados invertidos
        plt.scatter(x_data, y_data, label=f"{titulo}", color=cor)

        # Plotar a linha de ajuste linear
        plt.plot(x_data, slope * x_data + intercept, color='black', linestyle='--', linewidth=1.5, label=f"Fit {titulo} (slope: {slope:.2f})")

        # Imprimir os parâmetros do ajuste linear
        print(f"Ajuste Linear para {titulo}: slope = {slope}, intercept = {intercept}")

        # Calcular a constante de Curie
        curie_constant = 1 / slope
        print(f"Constante de Curie para {titulo}: {curie_constant}")

        # Calcular a temperatura de Curie
        curie_temperature = abs(intercept / slope)
        print(f"Temperatura de Curie para {titulo}: {curie_temperature}")

        # Calcular o momento efetivo
        effective_moment = np.sqrt(8 * curie_constant)
        print(f"Momento Efetivo para {titulo}: {effective_moment}")

# Adicionar legendas e títulos
plt.xlabel('Temperature (K)')
plt.ylabel('1/chi (Oe×mol×emu-1)')
plt.title('Inverso de Chi e ajuste linear')
plt.legend()

# Adicionar linha de grade
plt.grid(True)

# Mostrar o gráfico
plt.show()
```
